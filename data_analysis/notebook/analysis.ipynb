```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# load dataset
df=pd.read_csv('../data/raw/sales_data.csv')
print(df.shape)
print(df.info())
print(df.describe())
df.isnull().sum()
print(df)
```

```python
# Data cleaning

df['Sales']=pd.to_numeric(df['Sales'])
df.dropna(inplace=True)
df['Month']=df['Month'].str.strip()
df.head()
```

```python
df['Dept'].unique()
df['City'].unique()
```

```python
dept_sales=df.groupby('Dept')['Sales'].sum()
dept_sales
```

```python
sns.barplot(x=dept_sales,y=dept_sales.values)
```

```python
top_emp=df.groupby("Name")['Sales'].sum().sort_values(ascending=False)
sns.barplot(x=top_emp,y=top_emp.values)
```

```python
monthly=df.groupby("Month")["Sales"].sum()
sns.lineplot(x=monthly.index,y=monthly.values,markers='o')
```
