```python
df=pd.read_csv('sales_data.csv')
```

```python
ram_data=df[df['Name']=="Ram"]
plt.plot(ram_data['Month'],ram_data['Sales'])
plt.title("Ram's Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
df['Sales']=df['Sales'].replace('400s','400')
df['Sales']=pd.to_numeric(df['Sales'])

dept_sales=df.groupby("Dept")["Sales"].sum()
plt.bar(dept_sales.index,dept_sales.values)
plt.title('Total Sales by Department')
plt.xlabel("Department")
plt.ylabel("Total Sales")
plt.show()
```

```python
for name in df['Name'].unique():
    data=df[df['Name'] == name]
    plt.plot(data['Month'],data['Sales'],label=name)

plt.title("Sales comparison")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

```python
feb=df[df['Month']=='Feb']
plt.bar(feb['Name'],feb['Sales'])
plt.title('February Sales')
plt.show()
```

```python
x=[1,2,3,4]
y=[10,20,5,25]

plt.subplot(2,2,1)
plt.plot(x,y)
plt.title("Line chart1")

b=[10,20,30,40]
plt.subplot(2,2,2)
plt.plot(x,b)
plt.title("Line chart2")

b=[10,20,30,40]
plt.subplot(2,2,3)
plt.plot(x,b)
plt.title("Line chart2")

b=[10,20,30,40]
plt.subplot(2,2,4)
plt.plot(x,b)
plt.title("Line chart2")

plt.suptitle("Dashboard")
plt.show()
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
ram_data=df[df['Name']=="Ram"]
plt.plot(ram_data['Month'],ram_data['Sales'])
plt.title("Ram's Sales Trend")


plt.subplot(2,2,2)
dept_sales=df.groupby("Dept")["Sales"].sum()
plt.bar(dept_sales.index,dept_sales.values)
plt.title('Dept Sales')

plt.subplot(2,2,3)
for name in df['Name'].unique():
    data=df[df['Name'] == name]
    plt.plot(data['Month'],data['Sales'],label=name)
plt.title('Dept Sales')
plt.legend()


plt.subplot(2,2,4)
plt.hist(df['Sales'])
plt.title("Sales Distribution")

plt.show()
```

```python
df=pd.DataFrame({

    "Name":["Ram","Shyam","Hari","Sita","Gita","Amit"],
    "Dept":["IT","HR","Finance","Finance","HR","IT"],
    "Sales":[200,150,300,350,180,400],
    "Age":[25,30,25,32,26,29]
})
df
```

```python
plt.scatter(df["Age"],df["Sales"])
for i in range(len(df)):
    plt.text(df["Age"][i],df["Sales"][i],df["Name"][i])
plt.title("Age VS Sales")
plt.xlabel("Age")
plt.ylabel("Sales")
plt.show()
```

```python
plt.hist(df["Sales"])
plt.show()
```

```python
dept_sales  = df.groupby("Dept")['Sales'].sum()
plt.pie(dept_sales, labels = dept_sales.index, autopct = "%1.1f%%")
plt.show()
```

```python
explode=[0.1,0.1,0]
colors=["gold","lightblue","lightgreen"]
plt.pie(dept_sales,labels=dept_sales.index,autopct="%1.1f%%",explode=explode,colors=colors)
plt.show()
```
