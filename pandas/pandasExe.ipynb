```python
# loc : it returns data based on label
# syntax:     df.loc[row_label,column_label]
print(df.loc[0])
```

```python
# display only specific data like name only
print(df.loc[0,"Name"])
```

```python
# displaying like data from 1-3 index
print(df.loc[1:3])    # 3 is inclusive
```

```python
print(df.loc[:,"Name"])     # single column
```

```python
# multiple column
print(df.loc[:,["Name","Age"]])      # ' : ' means display all rows
```

```python
# iloc : return data based on index
# suntax : df.iloc[row_index,column_index]
print(df.iloc[0])
```

```python
print(df.iloc[0,1])
```

```python
# displaying like data from 1-3 index
print(df.iloc[1:3])     # 3 is exclusive
```

```python
print(df.iloc[:,2])   # : means all rows and 2 means index 2 from column
```

```python
# Difference between loc and iloc
  # loc : uses labels , end index inclusive 
  # iloc : uses possitions , end index exclusive
```

```python
print(df.loc[df["Salary"]>45000])   # checking single condition
```

```python
# increasing salary
df.loc[df["Department"]=="IT","Salary"]+=5000
print(df)
```

```python
# update values
df.loc[df["Age"]>25,"Department"]="Senoir"
print(df)
```

```python
# Adding column using loc[]
df.loc[:,"Bonus"]=df["Salary"]*0.1
print(df)
```
