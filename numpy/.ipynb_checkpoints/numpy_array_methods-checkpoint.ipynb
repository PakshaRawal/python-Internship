```python
import numpy as np
```

```python
marks = np.array(
   [
    [20, 30, 40], # student1
    [10, 12,13],  # student2
    [25, 32, 46], # student3
   ]
)
```

```python
print(np.sum(marks, axis=1)) # row wise
print(np.sum(marks, axis=0)) # col wise
```

```python
print(np.mean(marks, axis=0))
print(np.mean(marks, axis=1))
```

```python
print(np.max(marks, axis=0))
```

```python
print(marks.mean(axis=0)) # avg marks of each student
print(marks.mean(axis=1)) # avg marks of each subject
```

```python
# boolean indexing mean selecting element using condition
array = np.array([10, 20, 30, 40, 45, 50])
print(array>25)
```

```python
print(array[array>25])
```

```python
print(array[(array>25) & (array<45)])
```

```python
print(array[(array>25) | (array>45)])
```

```python
array[array>20]=0
print(array)
```

```python
marks = np.array([32, 45, 67, 78, 90, 23, 12])
```

```python
print(marks[marks>25])
```

```python
print(len(marks[marks>25]))
```

```python
result = np.where(marks>=32, "Pass", "Fail")
print(result)
```

```python
arr = np.array(
    [
        [20, 10, 20],
        [30, 40, 50]
    ])
row_sum = arr.sum(axis=1)
print(row_sum)
print(row_sum>60)
print(row_sum[row_sum>60])
```
