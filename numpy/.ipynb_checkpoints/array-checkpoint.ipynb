```python
import numpy as np
```

```python
arr = np.array([1,2,3])
print(arr)
```

```python
arr = np.array([[1,5,6],[3,5,6]])
print(arr)
```

```python
arr = np.array([[[3,4,2],[2,6,9]],[[1,0,3],[2,5,3]]])
print(arr[0,1,2])
arr.ndim
```

```python
arr= np.array([[0,2],[4,5]], np.int8)
print(arr[1,1])
arr.ndim
print(arr.dtype)
```

```python
arr = np.array([1,4,5])
print(arr[2])
arr.ndim
print(arr.dtype)
```

```python
arr = np.array([
                [[3,4,2],[2,6,9]],
                [[1,0,3],[2,5,3]],
                [[22,32,21],[12,13,14]
                ]])
print(arr[2,1,2])
arr.ndim
```

```python
arr= np.zeros((5,2))
print(arr)
```

```python
arr=np.ones((2,4))
print(arr)
print(arr.shape)
```

```python
arr=np.arange(0,12)
print(arr)
```

```python
arr=np.arange(0,10,2)
print(arr)
```

```python
arr=np.arange(10,0,-2)
print(arr)
```

```python
arr=np.arange(0,1,0.3)
print(arr)
```

```python
# slicing
arr=np.array([11,12,13,14,14])
print(arr[1:4:2])
```

```python
arr=np.array([[12,13,14],[1,2,3],[4,5,6]])
print(arr[0:2,1:3])
print(arr.flatten()) #flattening
```

```python
# reshaping
arr=np.array([1,2,3,4,5,6])
arr.shape
new_arr=arr.reshape(2,3)
print(new_arr)
new_arr.shape
```
