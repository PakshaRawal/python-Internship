```python
arr = np.array([[[3,4,2],[2,6,9]],[[1,0,3],[2,5,3]]])
print(arr[0,1,2])
arr.ndim
```

```python
arr = np.array([1,4,5])
print(arr[2])
arr.ndim
print(arr.dtype)
```

```python
arr = np.array([
                [[3,4,2],[2,6,9]],
                [[1,0,3],[2,5,3]],
                [[22,32,21],[12,13,14]
                ]])
print(arr[2,1,2])
arr.ndim
```

```python
arr= np.zeros((5,2))
print(arr)
```

```python
arr=np.ones((2,4))
print(arr)
print(arr.shape)
```

```python
arr=np.arange(0,12)
print(arr)
```

```python
# slicing
arr=np.array([11,12,13,14,14])
print(arr[1:4:2])
```

```python
arr=np.array([[12,13,14],[1,2,3],[4,5,6]])
print(arr[0:2,1:3])
print(arr.flatten()) #flattening
```

```python
# reshaping
arr=np.array([1,2,3,4,5,6])
arr.shape
new_arr=arr.reshape(2,3)
print(new_arr)
new_arr.shape
```

```python
arr = np.array([2,3,4,5])
result = arr*2
add = (arr+3)
print(result)
print(add)
```

```python
arr1=np.array([12,3,4])
arr2=np.array([11,23,3])
print(arr1+arr2)
print(arr1*arr2)
print(arr1/arr2)
print(arr1**2)
```

```python
salary=np.array([20000,30000,50000,60000])
print(salary*2.10)
```

```python
# universal functions ufuncs

arr=np.array([4,9,16])
result=np.sqrt(arr)
print(result)
print(np.exp(arr))
```

```python
arr=np.array([9,3, 90, np.pi*30])
print(np.sin(arr))
print(np.cos(arr))
```

```python
# statistical operations
salary=np.array([2000,3000,1000,9000])
print(np.sum(salary))
print(np.mean(salary))
print(np.average(salary))
print(np.min(salary))
print(np.max(salary))
print(np.max(salary)-np.min(salary))
print(np.std(salary)) #SD
print(np.median(salary))
print(np.percentile(salary,20))
print(np.cumsum(salary))
print(np.cumprod(salary))
print(np.diff(salary))
```
